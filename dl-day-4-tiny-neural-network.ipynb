# Network Structure & Notation

We will build a simple network:

- **Input layer:** 2 features ($x_1, x_2$)  
- **Hidden layer:** 2 neurons ($h_1, h_2$)  
- **Output layer:** 1 neuron ($o_1$)  

Weights and biases:

- Hidden layer weights: $$W^{[1]} = \begin{bmatrix}0.2 & 0.4 \\ 0.5 & -0.3\end{bmatrix}$$  
- Hidden layer biases: $$b^{[1]} = \begin{bmatrix}0.1 \\ -0.2\end{bmatrix}$$  
- Output layer weights: $$W^{[2]} = \begin{bmatrix}0.3 & -0.2\end{bmatrix}$$  
- Output layer bias: $$b^{[2]} = 0.05$$  

Activation functions:
- Hidden layer: ReLU  
- Output layer: Sigmoid


# Forward Propagation Formulas

**Hidden layer:**  

$$
Z^{[1]} = W^{[1]} X + b^{[1]}
$$

$$
A^{[1]} = ReLU(Z^{[1]})
$$

Where $X = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$

**Output layer:**  

$$
Z^{[2]} = W^{[2]} A^{[1]} + b^{[2]}
$$

$$
A^{[2]} = \sigma(Z^{[2]}) \quad \text{(Sigmoid)}
$$

**Prediction:**  

$$
\hat{y} = A^{[2]}
$$


# Define Activation Functions (Code)

```python
import numpy as np

# ReLU
def relu(x):
    return np.maximum(0, x)

# Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

# Forward Propagation Implementation (Code)

```python
# Input features
X = np.array([1.0, 2.0])

# Hidden layer weights and biases
W1 = np.array([[0.2, 0.4],
               [0.5, -0.3]])
b1 = np.array([0.1, -0.2])

# Output layer weights and bias
W2 = np.array([0.3, -0.2])
b2 = 0.05

# Forward pass: Hidden layer
Z1 = np.dot(W1, X) + b1
A1 = relu(Z1)

# Forward pass: Output layer
Z2 = np.dot(W2, A1) + b2
A2 = sigmoid(Z2)

print("Z1 (hidden pre-activation):", Z1)
print("A1 (hidden activation):", A1)
print("Z2 (output pre-activation):", Z2)
print("A2 (output activation / prediction):", A2)
```

# Experiment: Change Weights & Biases (Code)

```python
# Try different weights and biases
W1_new = np.array([[0.6, -0.1],
                   [-0.2, 0.5]])
b1_new = np.array([0.0, 0.3])
W2_new = np.array([0.4, 0.1])
b2_new = -0.1

# Forward pass
Z1_new = np.dot(W1_new, X) + b1_new
A1_new = relu(Z1_new)
Z2_new = np.dot(W2_new, A1_new) + b2_new
A2_new = sigmoid(Z2_new)

print("New prediction:", A2_new)
```

# Visualize Hidden Layer Activations

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

plt.bar(['h1', 'h2'], A1, color='skyblue', label='Original Weights')
plt.bar(['h1', 'h2'], A1_new, color='salmon', alpha=0.5, label='New Weights')

plt.title("Hidden Layer Activations: Original vs New Weights")
plt.ylabel("Activation Value")
plt.legend()
plt.show()
```

- The bar chart tells a story:

>“After weight update, hidden neurons are firing differently. Sparse activations can become more dense, which affects output strength.”

- Hidden layer activations are directly interpretable indicators of how the network is “thinking.”

- Small changes in weights/biases can significantly shift which neurons dominate.

# Key Takeaways from Day 4

1. **2-layer network** = input layer → hidden layer → output layer  
2. **Forward propagation** is just matrix multiplication + bias + activation:  
   $$A^{[l]} = activation(W^{[l]} A^{[l-1]} + b^{[l]})$$  
3. ReLU in hidden layer helps learn non-linear patterns  
4. Sigmoid in output layer produces a value between 0 and 1  
5. Changing weights/biases directly changes neuron activations and predictions
