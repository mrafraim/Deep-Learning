# What Are Hyperparameters?

Hyperparameters are values set before training begins.

Examples:

- Learning rate  
- Batch size  
- Weight decay  
- Optimizer  
- Scheduler  

They control:

> the dynamics of learning, not the representation itself

Most performance gains come from:

✔ Data quality & augmentation  
✔ Hyperparameter tuning  

NOT architecture changes.

This is why production ML teams invest heavily in tuning pipelines.

# Batch Size ↔ Learning Rate Coupling

## Core Relationship

Batch size affects gradient noise:

Small batch:
- Noisy gradients  
- Better generalization  
- Requires smaller LR  

Large batch:
- Stable gradients  
- Faster computation
- Allows larger LR

## Linear Scaling Rule

$$
LR_{new} = LR_{base} \times \frac{Batch_{new}}{Batch_{base}}
$$

Example:

Batch = 32 → LR = 0.001  
Batch = 128 → LR = 0.004





## Hidden Tradeoff

Large batch training often leads to:

• Sharp minima  
• Worse generalization  

Small batch training:

• Noisy but finds flatter minima  
• Better test performance  

This is why many vision models still prefer moderate batch sizes.

# Weight Decay (L2 Regularization)

L2 regularization is a technique used to prevent overfitting by adding a penalty to the loss function that discourages large weights.

It forces the model to keep weights small and smooth, reducing reliance on any single feature.

Weight decay adds penalty:

$$
L_{total} = L_{data} + \lambda ||w||^2
$$

It discourages large weights.

Where:

- $w$ = model weights
- $\lambda$ = regularization strength (hyperparameter)

L2 adds a penalty proportional to the square of the weights:

- Large weights → heavily penalized
- Small weights → lightly penalized

Result:

> The model spreads importance across features instead of relying on a few strong weights.

## Real Effect (Important Insight)

Weight decay does NOT just reduce weight size.

It:

• Controls model complexity  
• Encourages smoother functions  
• Improves generalization  

## Adam vs AdamW (Critical)

In Adam:

Weight decay is coupled with gradients → incorrect behavior.

In AdamW:

Weight decay is decoupled → correct regularization.

👉 Always prefer AdamW for CNNs.

```python
optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=1e-3,
    weight_decay=1e-4
)
```

Here:

`weight_decay` = λ

## Practical Weight Decay Ranges

| Scenario | Weight Decay |
|----------|-------------|
| Small dataset | 1e-3 |
| Medium dataset | 1e-4 |
| Large dataset | 1e-5 |

Signal:

- If overfitting → increase weight decay  
- If underfitting → decrease it  

# Controlled Experiments vs Efficient Search
## Beginner Approach

Change one parameter at a time.

Pros:

✔ Clear cause-effect understanding  

Cons:

❌ Extremely slow  
❌ Misses interactions between parameters  

## Professional Approach

Use a hybrid strategy:

1. Coarse search (broad exploration)  
2. Narrow search (fine tuning)  
3. Controlled validation (confirm results)  

This balances speed and understanding.

## Example Strategy

**Phase 1: Coarse Search:**

Try:

LR = [1e-2, 1e-3, 1e-4]  
Weight decay = [1e-3, 1e-4]

**Phase 2: Narrow Search:**

Zoom into best region:

LR = [3e-4, 5e-4, 7e-4]

## Why This Works

Hyperparameters interact:

• LR affects optimal weight decay  
• Batch size affects LR  
• Scheduler interacts with both  

So isolated tuning is incomplete.

# Efficient CNN Tuning Workflow

**Step 1: Find Stable Learning Rate**

• Run short training (5–10 epochs)  
• Identify divergence or slow learning  


**Step 2: Tune Weight Decay**

• Increase if overfitting  
• Decrease if underfitting  


**Step 3: Adjust Batch Size + LR Together**

• Scale LR with batch size  


**Step 4: Add Scheduler**

• Cosine or ReduceLROnPlateau  

**Step 5: Final Long Training**  

• Use best configuration  

# Reading Training Curves (Critical Skill)

Good training:

- Training loss decreases smoothly  
- Validation loss decreases then stabilizes  

Overfitting:

- Training ↓↓↓  
- Validation ↑  

Underfitting:

- Both losses high and flat  

# Common Tuning Mistakes

• Changing too many parameters blindly  
• Ignoring validation curves  
• Using full training runs too early  
• Overfitting to validation set  
• Not fixing random seeds  

# Practical CNN Tuning Pipeline

1️⃣ Start with AdamW (LR = 1e-3)  
2️⃣ Run short experiment  
3️⃣ Adjust LR based on behavior  
4️⃣ Add weight decay  
5️⃣ Tune batch size + LR  
6️⃣ Add scheduler  
7️⃣ Run full training  

# Why Tuning Matters

Training is an optimization problem.

Hyperparameters define:

- optimization landscape  
- gradient behavior  
- convergence speed  

Bad tuning:

- Model never reaches good solution.

Good tuning:

- Same architecture → dramatically better performance.

# Key Takeaways from Day 43

- Hyperparameters control learning dynamics  
- Batch size and LR must be tuned together  
- Weight decay is critical for generalization  
- Pure “one-variable” tuning is too slow  
- Use coarse-to-fine search strategy  
- Always rely on validation curves, not guesswork
