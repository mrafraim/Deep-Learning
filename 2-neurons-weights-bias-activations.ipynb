# Recap: What is a Neuron?

A neuron receives inputs, multiplies them by weights, adds a bias, and passes the result through an activation function.

Mathematically:

$$
output = activation(w_1 x_1 + w_2 x_2 + ... + w_n x_n + b)
$$

Where:
- $x_1, x_2, ..., x_n$ are input features  
- $w_1, w_2, ..., w_n$ are weights  
- $b$ is the bias  
- $activation()$ is a non-linear function (like ReLU, Sigmoid)


### Weights: Importance of Inputs

Weights control **how much influence** each input has on the neuron output.

- Large weight → input has strong impact  
- Small weight → input has little impact  
- Negative weight → input inversely affects output  

Example:

$$
output = 2 \cdot x_1 + 0.1 \cdot x_2 + (-1) \cdot x_3 + 0.5
$$

- $x_1$ dominates → most important  
- $x_2$ negligible  
- $x_3$ has negative effect


### Bias: Shifting the Activation

Bias allows the neuron to **shift the activation function**, giving it more flexibility.

Example without bias:

$$
output = w_1 x_1 + w_2 x_2
$$

If all inputs are 0, output is 0, neuron cannot activate.  

With bias:

$$
output = w_1 x_1 + w_2 x_2 + b
$$

Now the neuron can activate even if inputs are zero.  
This is critical for learning complex patterns.


# Activation Functions: Making Neurons Non-linear

Without activation functions, neurons only compute **linear combinations**:

$$
output = w \cdot x + b
$$

A stack of linear neurons is still **linear**, no matter how deep.  

Activation functions introduce **non-linearity**, allowing the network to learn complex patterns.

Common activations:
1. **ReLU (Rectified Linear Unit)**

$$
ReLU(x) = max(0, x)
$$

2. **Sigmoid**: maps output to (0, 1)

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

3. **Tanh**: maps output to (-1, 1)

$$
tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$


# Visualizing Neuron with Weights & Bias

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

# Inputs
plt.text(0.2, 0.8, "x1", fontsize=14)
plt.text(0.2, 0.6, "x2", fontsize=14)
plt.text(0.2, 0.4, "x3", fontsize=14)

# Neuron
plt.text(0.6, 0.6, "Neuron\n(w*x + b)\nActivation", fontsize=12, bbox=dict(facecolor='lightblue'))

# Arrows
plt.arrow(0.3, 0.8, 0.25, -0.1, width=0.003)
plt.arrow(0.3, 0.6, 0.25, 0.0, width=0.003)
plt.arrow(0.3, 0.4, 0.25, 0.1, width=0.003)

plt.axis('off')
plt.title("Neuron with Inputs, Weights, Bias and Activation")
plt.show()
```

# Forward Propagation with Weights & Bias

```python
import numpy as np

# Inputs
x = np.array([2.0, -1.0, 3.0])

# Weights
w = np.array([0.5, -0.2, 0.3])

# Bias
b = 0.1

# Linear output
linear_output = np.dot(x, w) + b
linear_output
```

# Applying Different Activation Functions

```python
# ReLU
def relu(x):
    return max(0, x)

# Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Tanh
def tanh(x):
    return np.tanh(x)

print("Linear output:", linear_output)
print("ReLU output:", relu(linear_output))
print("Sigmoid output:", sigmoid(linear_output))
print("Tanh output:", tanh(linear_output))
```

# Activation Function Visualization

```python
x_values = np.linspace(-5, 5, 100)

plt.figure(figsize=(10,4))

# ReLU
plt.subplot(1,3,1)
plt.plot(x_values, np.maximum(0, x_values))
plt.title("ReLU")
plt.grid(True)

# Sigmoid
plt.subplot(1,3,2)
plt.plot(x_values, 1 / (1 + np.exp(-x_values)))
plt.title("Sigmoid")
plt.grid(True)

# Tanh
plt.subplot(1,3,3)
plt.plot(x_values, np.tanh(x_values))
plt.title("Tanh")
plt.grid(True)

plt.show()
```

# Key Takeaways from Day 2

1. **Weights** determine input importance.  
2. **Bias** allows flexibility in neuron activation.  
3. **Activation functions** add non-linearity → critical for learning complex patterns.  
4. ReLU, Sigmoid, and Tanh are the most common; each has pros and cons.  
5. Forward propagation = $output = activation(\sum_i w_i x_i + b)$  
6. Visualization helps understand neuron behavior intuitively.
