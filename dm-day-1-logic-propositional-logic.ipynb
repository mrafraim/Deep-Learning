# Propositions and Truth Values

A proposition is a declarative statement that is either true (T) or false (F), but not both. It cannot be both and cannot be neither.

**Examples:**  
- "The sky is blue." → True (T)  
- "2 + 2 = 5" → False (F)  
- "x + 2 = 7" → Not a proposition (depends on x)

## Truth Values
- **T (True)** → The statement is correct
- **F (False)** → The statement is incorrect

Truth values form the foundation of:
- logical reasoning
- truth tables
- mathematical proofs
- digital logic circuits

## Classification of Sentences

| Type of Sentence | Proposition? | Reason |
|-----------------|-------------|--------|
| Declarative |  Yes | Has a definite truth value |
| Question | No | No truth value |
| Command | No | No truth value |
| Opinion | No | Subjective |
| Open statement | No | Depends on variable |

## Examples of Propositions

### Mathematical Statements
- "$7$ is a prime number." → **T**
- "$10$ is divisible by $3$." → **F**
- "$2^3 = 8$" → **T**

### Scientific / Real-World Statements
- "Water boils at $100^\circ C$ at sea level." → **T**
- "The sun revolves around the earth." → **F**

### Logical Statements
- "If $n$ is even, then $n^2$ is even." → **T**
- "All prime numbers are odd." → **F** (since $2$ is prime)

## Statements That Are NOT Propositions

### 1️. Open Statements
Statements containing variables without assigned values.

- "$x + 2 = 7$"
- "$n$ is a prime number"
- "$x > 0$"

These are not propositions because their truth depends on the value of the variable.

### 2️. Questions
- "Is $5$ a prime number?"
- "What time is it?"

Questions have no truth value.

### 3️. Commands
- "Close the door."
- "Solve the equation."

Commands have no truth value.

### 4️. Opinions / Subjective Statements
- "Mathematics is difficult."
- "This movie is interesting."

Opinions depend on personal belief, so they are not propositions.

## Converting Open Statements into Propositions

### Assigning a Value
- "$x + 2 = 7$" → Not a proposition (depends on x)
- If $x = 5$, then "$5 + 2 = 7$" → **T**

### Using Quantifiers
- "For all integers $n$, $n^2 \ge 0$" → **T**
- "There exists an integer $n$ such that $n^2 = -1$" → **F**

# Logical Operators

Logical operators are used to **combine** or **modify** propositions.

Let:
- P = First proposition
- Q = Second proposition

## 1. AND (Conjunction)

**Symbol:** ∧  
**Name:** Conjunction  
**Form:** P ∧ Q  

**Meaning:**  
True only when **both P and Q are true**.

### Truth Table

| P | Q | P ∧ Q |
|---|---|-------|
| T | T | T |
| T | F | F |
| F | T | F |
| F | F | F |

**Key Insight:**
AND is strict, one false makes everything false.

## 2. OR (Disjunction)

**Symbol:** ∨  
**Name:** Disjunction  
**Form:** P ∨ Q  

**Meaning:**  
True when **at least one** of P or Q is true.

### Truth Table

| P | Q | P ∨ Q |
|---|---|-------|
| T | T | T |
| T | F | T |
| F | T | T |
| F | F | F |

**Important:**
This is **inclusive OR**, not exclusive OR.

## 3. NOT (Negation)

**Symbol:** ¬  
**Form:** ¬P  

**Meaning:**  
Reverses the truth value.

### Truth Table

| P | ¬P |
|---|----|
| T | F |
| F | T |

**Double Negation Law:**

$$
\neg(\neg P) = P
$$


## 4. IMPLICATION (Conditional)

**Symbol:** →  
**Form:** P → Q  
Read as: "If P, then Q"

**Meaning:**  
False **only when P is true and Q is false**.

### Truth Table

| P | Q | P → Q |
|---|---|-------|
| T | T | T |
| T | F | F |
| F | T | T |
| F | F | T |

**Critical Insight:**

When P is false, the implication is **always true**.  
This is called **vacuous truth**.

Example:
"If 2 is odd, then 5 is prime." → True  
(Because the hypothesis is false.)

## 5. BICONDITIONAL (Equivalence)

**Symbol:** ↔  
**Form:** P ↔ Q  
Read as: "P if and only if Q"

**Meaning:**  
True when P and Q have the **same truth value**.

### Truth Table

| P | Q | P ↔ Q |
|---|---|-------|
| T | T | T |
| T | F | F |
| F | T | F |
| F | F | T |

**Equivalent Definition:**
$$
P \leftrightarrow Q \equiv (P \rightarrow Q) \land (Q \rightarrow P)
$$


## 6. Exclusive OR (XOR)

**Symbol:** ⊕  
**Name:** Exclusive OR  
**Form:** P ⊕ Q  

P ⊕ Q is **true if exactly one of P or Q is true**. It is **false if both are true or both are false**.

### Truth Table

| P | Q | P ⊕ Q |
|---|---|--------|
| T | T | F |
| T | F | T |
| F | T | T |
| F | F | F |

### Intuition

- OR (∨) → at least one true  
- XOR (⊕) → exactly one true  

Example:

Let  
P: "It is raining."  
Q: "It is sunny."

P ⊕ Q means:
"It is raining or sunny, but not both."

### Equivalent Logical Form

XOR can be expressed using basic operators:

$$
P \oplus Q \equiv (P \lor Q) \land \neg(P \land Q)
$$

OR

$$
P \oplus Q \equiv (P \land \neg Q) \lor (\neg P \land Q)
$$

## Common Exam Trap

Students confuse:

- $P \lor Q$ → allows both true  
- $P \oplus Q$ → forbids both true  

If both P and Q are true:
- OR → True  
- XOR → False  

## High-Yield Logical Identities (Very Important for Exams)

### 1️. De Morgan’s Laws

$$
\neg (P \land Q) \equiv \neg P \lor \neg Q
$$

$$
\neg (P \lor Q) \equiv \neg P \land \neg Q
$$


### 2️. Implication Conversion

$$
P \rightarrow Q \equiv \neg P \lor Q
$$

(Used heavily in proofs and simplifications.)


### 3️. Tautology & Contradiction Examples

Tautology (always true):
$$
P \lor \neg P
$$

Contradiction (always false):
$$
P \land \neg P
$$

# Tautology, Contradiction, and Logical Equivalence

These concepts are central to:
- Truth table analysis
- Logical simplification
- Proof techniques
- Boolean algebra
- Digital logic

## 1. Tautology

A **tautology** is a compound proposition that is true for all possible truth values of its variables.

### Example 1

$$
P \lor \neg P
$$

**Truth Table**

| P | ¬P | P ∨ ¬P |
|---|----|--------|
| T | F  | T |
| F | T  | T |

Since the final column is always **T**, it is a **tautology**.

### Example 2 (Important for Exams)

$$
(P \rightarrow Q) \lor (Q \rightarrow P)
$$

This is always true regardless of truth values.

### How to Prove Tautology

1. Construct a truth table  
2. If final column = all T → Tautology  

OR  

Use logical identities to reduce to **T**

## 2. Contradiction

A **contradiction** is a compound proposition that is **false for all possible truth values**.

### Example

$$
P \land \neg P
$$

**Truth Table**

| P | ¬P | P ∧ ¬P |
|---|----|--------|
| T | F  | F |
| F | T  | F |

Final column is always **F**, so it is a **contradiction**.


### Another Example

$$
\neg (P \lor \neg P)
$$

Since $(P \lor \neg P)$ is always true, its negation is always false.

## 3. Contingency (Very Important Concept)

A **contingency** is a proposition that is:
- True for some cases
- False for some cases

Example:

$$
P \land Q
$$

This is neither always true nor always false.

## 4. Logical Equivalence

Two propositions P and Q are **logically equivalent** if they have identical truth values for all cases.

### Symbol
$$
P \equiv Q
$$

This does NOT mean equality, 
It means **same truth table behavior**.


### Example: De Morgan’s Law

$$
\neg (P \land Q) \equiv \neg P \lor \neg Q
$$

**Truth Table Proof**

| P | Q | P ∧ Q | ¬(P ∧ Q) | ¬P | ¬Q | ¬P ∨ ¬Q |
|---|---|-------|----------|----|----|----------|
| T | T | T | F | F | F | F |
| T | F | F | T | F | T | T |
| F | T | F | T | T | F | T |
| F | F | F | T | T | T | T |

Columns for:
- ¬(P ∧ Q)
- ¬P ∨ ¬Q  

are identical → Therefore equivalent.


### High-Yield Equivalences (Must Memorize)

**1️. Implication Law**
                                                                                         
$$
P \rightarrow Q \equiv \neg P \lor Q
$$

**2️. Double Negation**
                                                                                         
$$
\neg(\neg P) \equiv P
$$

**3️. Biconditional**
                                                                                         
$$
P \leftrightarrow Q \equiv (P \rightarrow Q) \land (Q \rightarrow P)
$$

**4️. De Morgan’s Laws**
                                                                                         
$$
\neg (P \land Q) \equiv \neg P \lor \neg Q
$$

$$
\neg (P \lor Q) \equiv \neg P \land \neg Q
$$

## Relationship Summary

| Type | Always True? | Always False? |
|------|-------------|--------------|
| Tautology | ✅ Yes | No |
| Contradiction | No | ✅ Yes |
| Contingency | No | No |
